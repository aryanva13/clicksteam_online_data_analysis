```python
def utils_recognize_type(df, col, max_cat=20):
    if (df[col].dtype == "O") | (df[col].nunique() < max_cat):
        return "cat"
    else:
        return "num"
dic_cols = {col:utils_recognize_type(df, col, max_cat=20) for col in df.columns}
print(dic_cols)
```

```python
target=df['price 2']
df=df.drop('price 2',axis=1)
```

##**SPLITTING THE DATASET INTO TRAINING AND TESTING DATA**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

##**SCALING THE DATA USING STANDARD SCALER**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
col=['price']
X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

##**IMPLEMENTING PRINCIPAL COMPONENT ANALYSIS**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components =7) #reducing number of fetaures to 7
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
print(X_train)
```

##**IMPLEMENTING KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error,accuracy_score
```

```python
acc={}
from sklearn.model_selection import train_test_split, cross_val_score
k_values = list(range(1, 15))
mean_accuracy_scores = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    accuracy_scores = cross_val_score(knn, X_train, y_train, cv=5)  
    mean_accuracy_scores.append(np.mean(accuracy_scores))
optimal_k = k_values[np.argmax(mean_accuracy_scores)]
print("Mean Accuracy Scores:")
for k, accuracy in zip(k_values, mean_accuracy_scores):
    print(f"k = {k}: Accuracy = {accuracy:.4f}")
print("Optimal k:", optimal_k)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
acc_knn=accuracy_score(y_test,y_pred)
acc['knn']=acc_knn
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12, 6))
plt.plot(range(1, 15), mean_accuracy_scores, color='red', linestyle='dashed', marker='o',markerfacecolor='blue', markersize=10)
plt.title('K Value Accuracy Scores')
plt.xlabel('K Value')
plt.ylabel('Accuracy Score')
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
import seaborn as sns
cm=confusion_matrix(y_test, y_pred)
print(cm)
classes_names = ['price 1','price 2',]
cm = pd.DataFrame(confusion_matrix(y_test, y_pred),columns=classes_names, index = classes_names)
sns.heatmap(cm, annot=True, fmt='d');
```

```python
from yellowbrick.classifier import ROCAUC
def plot_ROC_curve(model, xtrain, ytrain, xtest, ytest):
    visualizer = ROCAUC(model, encoder={0: 'price 1', 
                                        1: 'price 2'})                                   
    visualizer.fit(xtrain, ytrain)
    visualizer.score(xtest, ytest)
    visualizer.show()
    return visualizer
plot_ROC_curve(knn, X_train,y_train,X_test, y_test)
```
